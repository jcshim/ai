## plt는 BGB을 RGB로, 2개의 영상 출력하기

```
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("jcshim.jpg")
img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
f, arr = plt.subplots(1,2)
arr[0].imshow(img)
arr[1].imshow(img_cvt)
#plt.show()
```
